```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
data1 = np.loadtxt('./result/total_reward/bi_ptr_reward.txt')
data2 = np.loadtxt('./result/total_reward/bi_rdn_reward.txt')
plt.style.use('fivethirtyeight')
plt.plot(range(len(data1)),data1[:,1], label='GNN+RL')
plt.plot(range(len(data2)),data2[:,1], label='RL')
plt.xlabel('epoch')
plt.ylabel('total reward')
plt.legend()
plt.savefig('./result/graph/compare21.jpg', dpi=200)
plt.show()
```

```python
data1 = np.loadtxt('./result/heuristic/bi_degree_reward_10_10.txt')
data2 = np.loadtxt('./result/heuristic/bi_CI_reward_10_10.txt')
data3 = np.loadtxt('./result/test/bi_result_rdn.txt')
data4 = np.loadtxt('./result/test/bi_result_ptr.txt')
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
plt.plot(range(len(data1)),data1, label='degree')
plt.plot(range(len(data2)),data2, label='CI')
plt.plot(range(len(data3)),data3, label='RL')
plt.plot(range(len(data4)),data4, label='GNN+RL')
plt.scatter(range(len(data1)),data1)
plt.scatter(range(len(data2)),data2)
plt.scatter(range(len(data3)),data3)
plt.scatter(range(len(data4)),data4)
plt.xlabel('num nodes')
plt.ylabel('reward')
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.tight_layout()
plt.legend()
plt.savefig('./result/graph/compare2.jpg', dpi=200)
plt.show()
```
