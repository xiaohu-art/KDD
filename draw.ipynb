```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

data1 = np.loadtxt('tra_result_rdn.txt')
data2 = np.loadtxt('tra_result_ptr.txt')
data3 = np.loadtxt('data/road/ANC_CI.txt')
data4 = np.loadtxt('data/road/ANC_degree.txt')
x = range(21)
plt.plot(x[1:],data1[:20,1],label = 'random feature')
plt.plot(x[1:],data2[:20,1],label = 'pretrain feature')
plt.plot(x[1:],data3[:20],label = 'CI')
plt.plot(x[1:],data4[:20],label = 'degree')
plt.style.use('fivethirtyeight')
plt.xlabel('node num')
plt.ylabel('ANC')
plt.legend()
plt.tight_layout()
plt.savefig('data/road/compare_anc_3.jpg',dpi = 200)
plt.show()

```
